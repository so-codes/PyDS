```python
import matplotlib.pyplot as plt
import numpy as np

# Given data
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86])

# Calculate the slope and intercept
mean_x = np.mean(x)
mean_y = np.mean(y)
numerator = np.sum((x - mean_x) * (y - mean_y))
denominator = np.sum((x - mean_x) ** 2)
slope_manual = numerator / denominator
intercept_manual = mean_y - slope_manual * mean_x

# Generate predicted values using the manual regression line
y_pred_manual = slope_manual * x + intercept_manual

# Plot the data and the manually fitted regression line
plt.scatter(x, y, label='Data')
plt.plot(x, y_pred_manual, color='green', label='Manual Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Manual Linear Regression')
plt.legend()
plt.show()
```
