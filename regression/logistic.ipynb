```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
x = np.array([5, 7, 8, 7, 2, 17, 2, 9, 4, 11, 12, 9, 6])
y = np.array([99, 86, 87, 88, 111, 86, 103, 87, 94, 78, 77, 85, 86])

# Define the logistic function
def manual_logistic_function(x, a, b):
    return 1 / (1 + np.exp(-a * (x - b)))

# Manually estimate parameters (you can adjust these values)
a_manual = 0.2
b_manual = 5

# Generate predicted values using the manual logistic function
y_pred_manual = manual_logistic_function(x, a_manual, b_manual)

# Plot the data and the manually fitted logistic curve
plt.scatter(x, y, label='Data')
plt.plot(x, y_pred_manual, color='green', label='Manual Logistic Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Manual Logistic Regression')
plt.legend()
plt.show()
```
