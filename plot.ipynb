```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 4, 9, 16, 25])
plt.plot(x, y)
# plt.plot(x, y,'o') # plot with dots
plt.show()

```

```python
# marker
# plt.plot(x, y, marker='o', color='r', linestyle='--', linewidth=2, markersize=10)

# string
plt.plot(x, y, '*--r') # markerlinecolor
plt.show()
```

```python
# Subplot
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 4, 9, 16, 25])
plt.subplot(1, 2, 1)
# plt.subplot(2, 1, 1)
plt.plot(x,y)
plt.title('Sales')


a = np.array([6,7,8,9,10])
b = np.array([20,30,21,15,10])
plt.subplot(1, 2, 2)
# plt.subplot(2, 1, 2)
plt.plot(a,b)
plt.title('Income')
plt.suptitle('ninad')
plt.grid()
plt.show()
```

```python
# Bargraph
import pandas as np
import matplotlib.pyplot as plt
x = np.array([1, 2, 3, 4, 5])
y = np.array([1, 4, 9, 16, 25])
# plt.bar(x, y)
plt.barh(x, y)
plt.show()
```

```python
# Histogram
x = np.random.normal(170, 10, 250)
plt.hist(x)
plt.show()
```

```python
y = np.array([1, 4, 9, 16, 25])
# plt.pie(data,explode,labels,colors,shadow)
plt.pie(y, labels=['a','b','c','d','e'], colors=['r','g','b','c','m'], shadow=True, explode=[0,0.5,0,0,0])
plt.legend(title = "Four Fruits:")
plt.show()
```
